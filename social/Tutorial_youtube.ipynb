### Cos'è un API?
<p>
Un API (Interfaccia di Programmazione delle Applicazioni) è un insieme di regole e protocolli che permette a diverse applicazioni software di comunicare tra loro. Serve come intermediario che consente a un'applicazione di richiedere e scambiare dati con un'altra, spesso in modo standardizzato e ben definito, semplificando l'integrazione e l'interazione tra sistemi diversi.
</p>

### Cos'è Oauth?
<p>
OAuth è un protocollo che permette a un'applicazione di terze parti di accedere a determinate informazioni o risorse in un servizio online a nome di un utente senza richiedere le credenziali di accesso dirette dell'utente. Quindi, quando un'applicazione di terze parti desidera accedere a dati su un sito web attendibile (come Gmail), OAuth consente di farlo senza dover condividere direttamente il nome utente e la password dell'utente con l'applicazione di terze parti.
</p>

```python
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
import json
import re
```

<p>
Recupero credenziali
</p>

<p>
Questo codice Python utilizza le librerie google_auth_oauthlib e googleapiclient per ottenere l'autorizzazione dall'utente e accedere all'API di YouTube. Le credenziali vengono caricate da un file JSON e, dopo l'autenticazione, l'oggetto youtube può essere utilizzato per interagire con l'API di YouTube, come ad esempio per recuperare informazioni sui video o sulle playlist
</p>

```python
scopes = ["https://www.googleapis.com/auth/youtube"]

client_credentials = "token/credenziali_s.json"
flow = InstalledAppFlow.from_client_secrets_file(client_credentials, scopes)
credentials = flow.run_local_server()
youtube=build('youtube','v3',credentials=credentials)
```

<p>
Estrai informazioni sul tuo canale:
</p>

<p>
Questo codice esegue una richiesta all'API di YouTube per ottenere le informazioni sul canale associato alle credenziali autorizzate.
La chiamata youtube.channels().list() con i parametri part="snippet" e mine=True richiede il recupero delle informazioni di tipo "snippet" relative al proprio canale.
La funzione execute() esegue effettivamente la richiesta all'API e restituisce le informazioni richieste.
</p>

```python
channel_info = youtube.channels().list(part="snippet", mine=True).execute()
print(json.dumps(channel_info, indent=2))
```

<p>
Estrai dettagli dei video con "like":
</p>

<p>
Il codice richiede all'API di YouTube di restituire i video che l'utente autenticato ha contrassegnato come "like" utilizzando i parametri specificati. Successivamente, stampa le informazioni di questi video in formato JSON per mostrare dettagli come titolo, descrizione e data di pubblicazione.
</p>

```python
popular_video = youtube.videos().list(part="snippet", myRating="like").execute()
print(json.dumps(popular_video, indent=2))
```

<p>
Estrai l'elenco dei video caricati nel tuo canale:

Questo codice utilizza l'API di YouTube per ottenere l'elenco delle sottoscrizioni dell'utente autenticato. <p>
Estrai i video correlati a 'python programming'
</p>

<p>
La funzione get_video() ricerca su YouTube video correlati a "python programming" e stampa i titoli dei primi 10 video insieme ai nomi dei canali che li hanno pubblicati.
</p>

```python
def get_video():
    response = youtube.search().list(
        part='snippet',
        maxResults=10,
        q='python programming'
    ).execute()

    for item in response['items']:
        print(f"Video: {item['snippet']['title']}")
        print(f"Canale: {item['snippet']['channelTitle']}\n")

get_video()
```

<p>
Estrai numero di visualizzazioni e mi piace di un video
</p>

<p>
Questa funzione, get_video_statistics(youtube, video_id), utilizza l'oggetto youtube per ottenere e stampare le statistiche di un video specificato mediante il suo ID. Attualmente, estrae e mostra le visualizzazioni e i "mi piace" del video identificato dall'ID 'NpWZ4_eQ9hY'. La riga relativa al conteggio dei "non mi piace" è commentata.
</p>

```python
def get_video_statistics(youtube, video_id):
    video_statistics = youtube.videos().list(
        part='statistics',
        id=video_id
    ).execute()

    statistics = video_statistics['items'][0]['statistics']
    print(f"Visualizzazioni: {statistics['viewCount']}")
    print(f"Mi piace: {statistics['likeCount']}")

get_video_statistics(youtube, 'NpWZ4_eQ9hY')
```

### Dataset a partire dal canale YouTube

Questo script utilizza l'**API di YouTube** per raccogliere dettagli su 500 video di un canale specifico. Fa delle richieste per ottenere i dettagli dei video e le rispettive statistiche, come visualizzazioni, mi piace e commenti, poi crea un DataFrame usando Pandas in Python per organizzare e presentare questi dati in forma tabellare

```python
channel_id = 'UCBR8-60-B28hp2BmDPdntcQ'
videos_data = []

next_page_token = None

while True: 
    videos = youtube.search().list(
        part='snippet',
        channelId=channel_id,
        type='video',
        maxResults=50,
        pageToken=next_page_token
    ).execute()

    total_videos = videos.get('items', [])

    for video in total_videos:
        video_id = video['id']['videoId']
        video_stats = youtube.videos().list(
            part='snippet,statistics',
            id=video_id
        ).execute()

        if 'items' in video_stats and video_stats['items']:
            snippet = video_stats['items'][0]['snippet']
            statistics = video_stats['items'][0]['statistics']

            title = snippet.get('title', None)
            published_at = snippet.get('publishedAt', None)
            category = snippet.get('categoryId', None)
            tags = snippet.get('tags', [])

            views = statistics.get('viewCount', 0)
            likes = statistics.get('likeCount', 0)
            comments = statistics.get('commentCount', 0)

            video_data = {
                'VideoID': video_id,
                'Title': title,
                'PublishedAt': published_at,
                'Category': category,
                'Tags': tags,
                'Views': views,
                'Likes': likes,
                'Comments': comments,
            }
            videos_data.append(video_data)
        else:
            print(f"Info mancanti per il video con ID: {video_id}")

    next_page_token = videos.get('nextPageToken')

    if not next_page_token:
        break

pd.set_option('display.max_rows', None)
df = pd.DataFrame(videos_data)
df
```

#### Dataset tendenza YouTube Italia

Questo codice Python utilizza l'**API di YouTube** per raccogliere informazioni sui video più popolari in Italia. Recupera dettagli come titolo, data di pubblicazione, categorie e metriche di coinvolgimento come visualizzazioni, like e commenti per ciascun video.

#### Recupero informazioni canali

Questa funzione, get_channel_stats, si occupa di raccogliere informazioni statistiche sui canali YouTube tramite l'uso dell'API di YouTube

```python
def get_channel_stats(youtube, channel_ids):
    all_data = []
    request = youtube.channels().list(
                part='snippet,contentDetails,statistics',
                id=','.join(channel_ids)
                )
    response = request.execute() 
    
    for i in range(len(response['items'])):
        data = dict(Channel_name = response['items'][i]['snippet']['title'],
                    Subscribers = response['items'][i]['statistics']['subscriberCount'],
                    Views = response['items'][i]['statistics']['viewCount'],
                    Total_videos = response['items'][i]['statistics']['videoCount'],
                    playlist_id = response['items'][i]['contentDetails']['relatedPlaylists']['uploads'])
        all_data.append(data)
    
    return all_data
```

```python
channel_statistics = get_channel_stats(youtube, channel_ids)
channel_data = pd.DataFrame(channel_statistics)
channel_data
```

#### Recupero informazioni sui video specifici

Questa funzione Python, get_video_ids, si occupa di recuperare gli ID dei video presenti in una specifica playlist su YouTube tramite l'utilizzo dell'API di YouTube

```python
channel_data['Subscribers'] = pd.to_numeric(channel_data['Subscribers'])
channel_data['Views'] = pd.to_numeric(channel_data['Views'])
channel_data['Total_videos'] = pd.to_numeric(channel_data['Total_videos'])
```

```python
def get_video_ids(youtube, playlist_id):
    
    request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId = playlist_id,
                maxResults = 50)
    response = request.execute()
    
    video_ids = []
    
    for i in range(len(response['items'])):
        video_ids.append(response['items'][i]['contentDetails']['videoId'])
        
    next_page_token = response.get('nextPageToken')
    more_pages = True
    
    while more_pages:
        if next_page_token is None:
            more_pages = False
        else:
            request = youtube.playlistItems().list(
                        part='contentDetails',
                        playlistId = playlist_id,
                        maxResults = 50,
                        pageToken = next_page_token)
            response = request.execute()
    
            for i in range(len(response['items'])):
                video_ids.append(response['items'][i]['contentDetails']['videoId'])
            
            next_page_token = response.get('nextPageToken')
        
    return video_ids
```

```python
playlist_id = channel_data.loc[channel_data['Channel_name']=='Hamza Ahmed', 'playlist_id'].iloc[0]
video_ids = get_video_ids(youtube, playlist_id)
video_ids
```

```python
def get_video_details(youtube, video_ids):
    all_video_stats = []
    
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
                    part='snippet,contentDetails,statistics',
                    id=','.join(video_ids[i:i+50]))
        response = request.execute()
        
        for video in response['items']:
            video_stats = {
                'Title': video['snippet']['title'],
                'Published_date': video['snippet']['publishedAt'],
                'Views': video['statistics']['viewCount'],
                'Likes': video['statistics']['likeCount'],
                'Comments': video['statistics']['commentCount'],
                'VideoCategory': video['snippet']['categoryId']
            }
            
            if 'contentDetails' in video and 'duration' in video['contentDetails']:
                video_stats['Duration'] = video['contentDetails']['duration']
            
            all_video_stats.append(video_stats)
    
    return all_video_stats

video_details = get_video_details(youtube, video_ids)
video_data = pd.DataFrame(video_details)
video_data
```

L'asse x rappresenta i tag, l'asse y rappresenta la frequenza di ciascun tag. Il grafico offre un'istantanea chiara dei tag più frequenti all'interno del dataset, aiutando a comprendere quali tematiche o argomenti siano più comuni nei video.

```python
top_tags = Counter([tag for sublist in video_for_channels['tags'] for tag in sublist]).most_common(10)
labels, values = zip(*top_tags)
plt.figure(figsize=(10, 6))
plt.bar(labels, values)
plt.title('Top 10 Tags')
plt.xlabel('Tags')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

Da questo grafico risulta che ci sono diversi tag abbastanza frequenti come news o music, anche questo risulta un aspetto importante per l'analisi soprattutto per capire più nello specifico quali sono gli argomenti più ricorrenti nei video.

#### Grafico a barre sulle visualizzazioni per paese

Questo codice utilizza Matplotlib per visualizzare le visualizzazioni totali dei video suddivise per paese attraverso un grafico a barre. Convertendo i dati e rimuovendo le voci 'Unknown' per i paesi, calcola le visualizzazioni totali per ciascun paese e le rappresenta visivamente. Il grafico permette di confrontare facilmente le visualizzazioni tra i diversi paesi nel dataset dei video.

```python
video_for_channels['views'] = video_for_channels['views'].astype(int)

filtered_data = video_for_channels[video_for_channels['country'] != 'Unknown']

views_by_country = filtered_data.groupby('country')['views'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
views_by_country.plot(kind='bar', color='skyblue')
plt.title('Visualizzazioni per Paese')
plt.xlabel('Paese')
plt.ylabel('Visualizzazioni')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Da questo grafico risulta che i video con il numero più alto di visualizzazioni sono stati caricati negli stati uniti, per cui anche il paese di provenienza risulta un fattore da tenere in considerazione per la nostra analisi.
